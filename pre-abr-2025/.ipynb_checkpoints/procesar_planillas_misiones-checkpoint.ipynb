```python
import os
import pandas as pd
import numpy as np
from glob import glob
```

```python
csv_list = glob("entrerios_test/*.csv")
```

```python
df = pd.read_csv(csv_list[0], skiprows=1)
```

```python
csv_list
```

```python
for csv in csv_list[1:]:
    df_tmp = pd.read_csv(csv, skiprows=1)
    df = pd.concat([df, df_tmp], ignore_index=True)
```

```python
# df.PRUEBA.map('{:,.0f}'.format).str.zfill(4)
```

```python
df.duplicated().sum()
```

```python
df = df.iloc[:, :-1]
```

```python
column_names = [
    "Código de anonimización",
    "Fecha de nacimiento",
    "Género",
    "Jardín",
    "Sala",
    "version_voc",
    "ID_voc_tablet",
    "ID_voc_prueba",
    "version_conc_fono",
    "ID_conc_fono_tablet",
    "ID_conc_fono_prueba",
    "version_sist_escrit",
    "ID_sist_escrit_tablet",
    "ID_sist_escrit_prueba",
    "Supervisión",
    "localidad",
    "Ocupación de los padres",
    "Comentarios"]
# Renombrar las columnas del DataFrame
df.columns = column_names
```

```python
df["ID_voc_prueba"] = (
    df["ID_voc_prueba"]
    .fillna(0)  # Opcional: manejar NaN (por ejemplo, reemplazarlos por 0)
    .astype(int)  # Asegurar que los valores son enteros
    .map('{:04d}'.format)  # Formatear con ceros a la izquierda
)
# Convertir a formato numérico y rellenar con ceros (manejo de NaN)
df["ID_conc_fono_prueba"] = (
    df["ID_conc_fono_prueba"]
    .fillna(0)  # Opcional: manejar NaN (por ejemplo, reemplazarlos por 0)
    .astype(int)  # Asegurar que los valores son enteros
    .map('{:04d}'.format)  # Formatear con ceros a la izquierda
)
df["ID_sist_escrit_prueba"] = (
    df["ID_sist_escrit_prueba"]
    .fillna(0)  # Opcional: manejar NaN (por ejemplo, reemplazarlos por 0)
    .astype(int)  # Asegurar que los valores son enteros
    .map('{:04d}'.format)  # Formatear con ceros a la izquierda
)
```

```python
df["ID_voc_tablet"] = df["ID_voc_tablet"].fillna(0).astype(int)
df["ID_conc_fono_tablet"] = df["ID_conc_fono_tablet"].fillna(0).astype(int)
df["ID_sist_escrit_tablet"] = df["ID_sist_escrit_tablet"].fillna(0).astype(int)
```

```python
df["ID_voc"] = df["ID_voc_tablet"].astype(str) + "-" + df["ID_voc_prueba"].astype(str)
df["ID_conc_fono"] = df["ID_conc_fono_tablet"].astype(str) + "-" + df["ID_conc_fono_prueba"].astype(str)
df["ID_sist_escrit"] = df["ID_sist_escrit_tablet"].astype(str) + "-" + df["ID_sist_escrit_prueba"].astype(str)
df
```

```python
df.drop(columns=["ID_voc_tablet", "ID_voc_prueba","ID_conc_fono_tablet",    "ID_conc_fono_prueba",     "ID_sist_escrit_tablet",    "ID_sist_escrit_prueba",    "ID_sist_escrit_prueba"], inplace=True)
```

```python
df.shape
```

```python
df.to_csv("df_entrerios.csv")
```

```python
df_melted = pd.melt(
    df,
    id_vars=[
        "Código de anonimización",
        "Fecha de nacimiento",
        "Género",
        "Jardín",
        "Sala",
        "version_voc",
        "version_conc_fono",
        "version_sist_escrit",
        "Supervisión",
        "localidad",
        "Ocupación de los padres",
        "Comentarios"
    ],
    value_vars=["ID_sist_escrit", "ID_conc_fono", "ID_voc"],
    var_name="prueba",
    value_name="test_id",
    col_level=None,
    ignore_index=True
)
```

```python
# Lo uso para descargar la data del server
df_melted["test_id"].to_csv("listado_ids_entrerios.csv") 
```
